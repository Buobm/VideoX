Skript to build Kinetics 400 dataset 

```python
#get from  labels/kinetics_400_labels.csv the labels and IDs out of the csv file and generate a dictionarry whare the key is the ID and the value is the label
import csv
import os
import pandas as pd
import numpy as np

def get_labels():
    labels = {}
    with open('labels/kinetics_400_labels.csv', 'r') as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        for row in reader:
            if row[0] == 'id':
                continue
            labels[row[1]] = row[0]
    return labels

labels = get_labels()
labels
```

```python
#function to load the data from the csv file and return an pandas array of the data using the first row as the header
def load_data(path):
    data = pd.read_csv(path, header=0)
    return data
data = load_data('kinetics-400/k400/annotations/test.csv')
data.head()
```

```python
#generate a .txt file where each line is first the video name and then the label
def generate_txt(txt_path, data_path):
    data = pd.read_csv(data_path, header=0)
    with open(txt_path, 'w') as txtfile:
        for index, row in data.iterrows():
            #get key from dictionary corresponding to the label
            class_label = labels[row['label']]
            start = str(row['time_start']).zfill(6)
            end = str(row['time_end']).zfill(6)
            txtfile.write(row['youtube_id'] + '_' + start + '_' + end + '.mp4' + ' ' + str(class_label) + '\n')

generate_txt('kinetics-400/k400/annotations/train.txt', 'kinetics-400/k400/annotations/train.csv')
```
